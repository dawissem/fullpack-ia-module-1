```python
'''2. Analyse des Tendances et DÃ©composition de la SÃ©rie'''
```

```python
'''6. DÃ©composer la sÃ©rie temporelle en tendance, saisonnalitÃ© et rÃ©sidu.'''
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series using seasonal decomposition
decomposition = seasonal_decompose(data2['Daily minimum temperatures'], model='additive', period=365)

# Plot the decomposition
plt.figure(figsize=(12, 8))

# Plot the observed data, trend, seasonality, and residuals
plt.subplot(411)
plt.plot(decomposition.observed, label='Observed', color='blue')
plt.title('Observed Data')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color='orange')
plt.title('Trend')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal', color='green')
plt.title('Seasonality')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residuals', color='red')
plt.title('Residuals')

# Adjust layout to make it clearer
plt.tight_layout()
plt.show()
```

```python
'''' Appliquer une moyenne mobile (7 jours et 30 jours) et comparer avec la sÃ©rie
originale.'''
```

```python
# Apply 7-day and 30-day moving averages to the 'Daily minimum temperatures' column
data2['7-day MA'] = data2['Daily minimum temperatures'].rolling(window=7).mean()
data2['30-day MA'] = data2['Daily minimum temperatures'].rolling(window=30).mean()

# Plot the original data along with the 7-day and 30-day moving averages
plt.figure(figsize=(12, 6))

# Plot the original data and moving averages
plt.plot(data2['Date'], data2['Daily minimum temperatures'], label='Original Data', color='blue', alpha=0.6)
plt.plot(data2['Date'], data2['7-day MA'], label='7-Day Moving Average', color='orange', linewidth=2)
plt.plot(data2['Date'], data2['30-day MA'], label='30-Day Moving Average', color='green', linewidth=2)

# Add labels and title
plt.title('Comparison of Original Data with 7-Day and 30-Day Moving Averages', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily Minimum Temperature (Â°C)', fontsize=12)
plt.legend()

# Adjust layout and display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
''' 8. Identifier les mois les plus froids et les plus chauds sur lâ€™ensemble des annÃ©es
disponibles. ''''
```

```python
# Extract the month from the 'Date' column
data2['Month'] = data2['Date'].dt.month

# Calculate the average temperature for each month
monthly_avg_temp = data2.groupby('Month')['Daily minimum temperatures'].mean()

# Identify the coldest and hottest months
coldest_month = monthly_avg_temp.idxmin()
hottest_month = monthly_avg_temp.idxmax()

# Get the average temperatures of the coldest and hottest months
coldest_month_temp = monthly_avg_temp.min()
hottest_month_temp = monthly_avg_temp.max()

# Display the results
coldest_month, coldest_month_temp, hottest_month, hottest_month_temp
```

```python
''' 9. Calculer lâ€™autocorrÃ©lation et tracer le graphique ACF (Autocorrelation Function). '''
```

```python
from statsmodels.graphics.tsaplots import plot_acf

# Calculate and plot the Autocorrelation Function (ACF)
plt.figure(figsize=(10, 6))
plot_acf(data2['Daily minimum temperatures'], lags=50, ax=plt.gca())
plt.title('Autocorrelation Function (ACF) of Daily Minimum Temperatures', fontsize=14)
plt.xlabel('Lag', fontsize=12)
plt.ylabel('Autocorrelation', fontsize=12)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
''''10. Identifier si la sÃ©rie est stationnaire en utilisant le test de Dickey-Fuller. '''
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the Dickey-Fuller test for stationarity
adf_result = adfuller(data2['Daily minimum temperatures'])

# Extract and display the p-value and test statistic
adf_statistic = adf_result[0]
adf_p_value = adf_result[1]
adf_critical_values = adf_result[4]

adf_statistic, adf_p_value, adf_critical_values
```
