```python
# SÃ©lection des colonnes numÃ©riques uniquement
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calcul de la matrice de corrÃ©lation
correlation_matrix = numeric_data.corr()

# Affichage de la matrice de corrÃ©lation sous forme de tableau interactif
#tools.display_dataframe_to_user(name="Matrice de CorrÃ©lation", dataframe=correlation_matrix)

# Identification de la paire de variables la plus corrÃ©lÃ©e (hors diagonale)
corr_unstacked = correlation_matrix.unstack()
corr_unstacked = corr_unstacked[corr_unstacked < 1]  # Exclure la diagonale (corrÃ©lation parfaite)
most_correlated_pair = corr_unstacked.abs().idxmax()  # Trouver la paire la plus corrÃ©lÃ©e
highest_correlation = corr_unstacked[most_correlated_pair]

most_correlated_pair, highest_correlation
```
