```python
# Sélection des colonnes numériques uniquement
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calcul de la matrice de corrélation
correlation_matrix = numeric_data.corr()

# Affichage de la matrice de corrélation sous forme de tableau interactif
#tools.display_dataframe_to_user(name="Matrice de Corrélation", dataframe=correlation_matrix)

# Identification de la paire de variables la plus corrélée (hors diagonale)
corr_unstacked = correlation_matrix.unstack()
corr_unstacked = corr_unstacked[corr_unstacked < 1]  # Exclure la diagonale (corrélation parfaite)
most_correlated_pair = corr_unstacked.abs().idxmax()  # Trouver la paire la plus corrélée
highest_correlation = corr_unstacked[most_correlated_pair]

most_correlated_pair, highest_correlation
```

```python
# Sélectionner uniquement les colonnes numériques et remplacer les NaN par 0
correlation_matrix = data.select_dtypes(include=['number']).fillna(0).corr()

# Identifier les paires de variables ayant une corrélation > 0.7 (hors diagonale)
correlation_threshold = 0.7
high_correlation_pairs = correlation_matrix.where((correlation_matrix.abs() > correlation_threshold) & (correlation_matrix.abs() < 1))

# Convertir en format lisible (liste des paires)
high_correlation_pairs = high_correlation_pairs.stack().reset_index()
high_correlation_pairs.columns = ["Variable 1", "Variable 2", "Correlation"]
high_correlation_pairs = high_correlation_pairs.sort_values(by="Correlation", ascending=False)
```
