```python
# üì¶ Installation des d√©pendances
!pip install -q langchain openai faiss-cpu gradio tiktoken unstructured pypdf

# üß† Initialisation
import os
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import PyPDFLoader, TextLoader
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
import gradio as gr

# üîê Cl√© API OpenAI (remplacer ici ou utiliser os.environ)
os.environ['OPENAI_API_KEY'] = ""  # ‚Üê Remplace par ta cl√©

# üìÇ Chargement de documents : uploader un ou plusieurs fichiers PDF ou .txt
from google.colab import files
uploaded = files.upload()  # uploader des fichiers docs de r√©f√©rence

docs = []
for fname in uploaded.keys():
    if fname.endswith(".pdf"):
        loader = PyPDFLoader(fname)
    else:
        loader = TextLoader(fname)
    docs.extend(loader.load())

# ‚úÇÔ∏è Chunking + vectorisation FAISS
splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=150)
chunks = splitter.split_documents(docs)

embeddings = OpenAIEmbeddings()
db = FAISS.from_documents(chunks, embeddings)

# ü§ñ Construction du chatbot avec RAG
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 3})
llm = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo")
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)

# üß™ Test simple
query = "Comment configurer le service ?"
response = qa_chain.run(query)
print("Exemple de r√©ponse:", response)

# üéõÔ∏è Interface Gradio
def chatbot_interface(question):
    return qa_chain.run(question)

gr.Interface(fn=chatbot_interface,
             inputs=gr.Textbox(lines=2, placeholder="Pose une question sur la documentation..."),
             outputs="text",
             title="Chatbot Support Technique",
             description="Assistant aliment√© par RAG (OpenAI + FAISS)").launch()
```
